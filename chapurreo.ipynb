```python
# Definir el mapeo de valores categóricos
mapeo = {
    1: 'sunny',
    2: 'a little bit cloudy',
    3: 'cloudy',
    4: 'very cloudy',
    5: 'stormy'
}

# Crear una lista vacía para almacenar los valores categóricos
weather_category = []

# Iterar sobre los valores numéricos de la columna weather_type y asignar los valores categóricos
for valor in df3['weather_type']:
    weather_category.append(mapeo.get(valor, 'desconocido'))

# Agregar la lista de valores categóricos al DataFrame como una nueva columna
df3['weather_category'] = weather_category

```

```python
df3.columns
```

```python
df3['weather_type'] = df3['weather_category']
df3 = df3[['Time', 'Energy delta[Wh]', 'GHI', 'temp', 'pressure', 'humidity', 'wind_speed', 'rain_1h', 'snow_1h', 'clouds_all', 'isSun', 
    'sunlightTime', 'dayLength', 'SunlightTime/daylength', 'weather_type',
       'hour', 'month']]

df3.head()
```
