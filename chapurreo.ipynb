```python
import pandas as pd
import numpy as np
from scipy import stats

def replace_outliers_with_mean_or_mode(df):
    for col in df.columns:
        # Calcula los cuartiles Q1 y Q3 para cada columna
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)

        # Calcula el rango intercuartílico (IQR) para cada columna
        iqr = q3 - q1

        # Filtra los valores entre Q1-1.5 * IQR y Q3+1.5 * IQR
        filtered = df.query('(@q1 - 1.5 * @iqr) <= {} <= (@q3 + 1.5 * @iqr)'.format(col))

        # Verifica la asimetría de la distribución
        skewness = df[col].skew()

        # Reemplaza los outliers
        if skewness > 0:  # Distribución asimétrica positiva (usamos la moda)
            mode_val = filtered[col].mode().iloc[0]
            df[col] = np.where((df[col] > q3 + 1.5 * iqr) | (df[col] < q1 - 1.5 * iqr), mode_val, df[col])
        else:  # Distribución normal (usamos la media)
            mean_val = filtered[col].mean()
            df[col] = np.where((df[col] > q3 + 1.5 * iqr) | (df[col] < q1 - 1.5 * iqr), mean_val, df[col])

    return df

# Ejemplo de uso con tu DataFrame 'no_outliers_data'
no_outliers_data_cleaned = replace_outliers_with_mean_or_mode(no_outliers_data)
```
